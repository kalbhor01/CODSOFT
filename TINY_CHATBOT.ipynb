```python
import re
import random
import textwrap
import datetime

RULES = [

    (r"\bhi+\b|\bhello+\b|\bhey+\b", [
        "hii",
        "Hello there! ",
        "Hi! How can I help you today?",
        "Hey! What's up?"
    ]),


    (r"\bhow (are|r) (you|u)\b", [
        "I'm just a bunch of Python code, but I'm feeling ",
        "Doing great, thanks for asking. What about you?"
    ]),


    (r"\b(date|time|day)\b",      lambda _: get_date_time()),


    (r"\b(weather|temp(erature)?)\b", lambda _: get_weather_stub()),


    (r"\bbye\b|\bexit\b|\bquit\b", "bye"),
]


DEFAULT_FALLBACKS = [
    "Sorry, I didnt get that. Could you rephrase?",
    "Hmm, Im not sure I understand.",
    "Interestingâ€¦ tell me more!",
]
RULES = [
    (r"\bhi\b|\bhello\b|\bhey\b", [...]),
    (r"\bhow (are|r) (you|u)\b", [...]),
    (r"\b(date|time|day)\b",      lambda _: get_date_time()),   # â† NEW
    (r"\b(weather|temp(erature)?)\b", lambda _: get_weather_stub()),
    (r"\bbye\b|\bexit\b|\bquit\b", "bye"),
]


def get_weather_stub() -> str:
    """
    Pretend to fetch weather data.
    Replace this stub with a real API call later if you like.
    """
    demo_conditions = ["sunny", "cloudy", "rainy", "stormy", "foggy"]
    return f"It looks {random.choice(demo_conditions)} where I am!"

def get_date_time() -> str:
    """
    Return todayâ€™s date and current local time in a friendly format.
    Works on Windows and Unixâ€‘like OSs.
    Example:  'Monday, 30 June 2025 â€¢ 02:45 PM'
    """
    now = datetime.datetime.now()

    # Get day of month without leading zero in a portable way
    day = now.day                      # 1â€‘31 as int
    formatted = now.strftime("%A, %B %Y â€¢ %I:%M %p")
    # Insert the raw day value after weekday and a comma
    # Monday, 30 June 2025 â€¢ 02:45 PM
    return f"{now.strftime('%A')}, {day} {now.strftime('%B %Y â€¢ %I:%M %p')}"

def match_rule(user_text: str):
    """
    Iterate through RULES and return:
        â€¢ reply           (str)
        â€¢ should_quit     (bool)  â€“ True if the user said bye/exit/quit
    """
    for pattern, response in RULES:
        if re.search(pattern, user_text, re.I):
            # If the ruleâ€™s response is a function, call it.
            if callable(response):
                return response(user_text), False

            # If itâ€™s a list, pick a random reply.
            if isinstance(response, list):
                return random.choice(response), False

            # If the sentinel string "bye" is stored, exit.
            if response == "bye":
                return "Goodâ€‘bye! ğŸ‘‹", True

    # No rule matched â†’ fall back politely
    DEFAULT_FALLBACKS = [
        "Sorry, I didnâ€™t get that. Could you rephrase?",
        "Hmm, Iâ€™m not sure I understand.",
        "Interestingâ€¦ tell me more!",
    ]
    return random.choice(DEFAULT_FALLBACKS), False

def chat():
    banner = textwrap.dedent("""
        â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
        â•‘      Tiny-Chatbot                          â•‘
        â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    """).strip()
    print(banner)

    while True:
        user = input("You : ").strip()
        if not user:
            continue

        reply, should_quit = match_rule(user)
        print("Bot :", reply)

        if should_quit:
            break

if __name__ == "__main__":
   try:
       chat()
   except KeyboardInterrupt:
       print("\nBot : Bye for now!")


```
